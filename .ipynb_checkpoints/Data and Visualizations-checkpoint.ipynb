```python

# Random Data Generation
np.random.seed(0)
n, p = 40, 8
d = np.random.normal(0, 2, (n, p))
d += np.log(np.arange(1, p + 1)) * -5 + 10


# plot
sns.set_style('ticks')
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)
sns.violinplot(data=d, inner="points", ax=ax)    
sns.despine()

```

```python

df=pd.DataFrame({'x': range(1,101), 'y': np.random.randn(100)*15+range(1,101), 
                 'z': (np.random.randn(100)*15+range(1,101))*2 })
 
# Creating a 2x2 subplot of various data makrers - with a real dataset these 
#suplots could represent any number of variables
plt.subplot(221)
plt.plot( 'x', 'y', data=df, marker='o', alpha=0.4)
plt.subplot(222)
plt.plot( 'x','z', data=df, linestyle='none', marker='o', color="orange", alpha=0.3)
plt.subplot(223)
plt.plot( 'x','z', data=df, linestyle='none', marker='D', color="cyan", alpha=0.3)
plt.subplot(224)
plt.plot( 'x','z', data=df, marker='x', color="grey", alpha=0.3)
plt.show()
```

```python
#free dataset I've used for another project
df = pd.read_csv("mtcars.csv")
x = df.loc[:, ['mpg']]
df['mpg_z'] = (x - x.mean())/x.std()

# Conditional function for plotting + OR - 
df['colors'] = ['red' if x < 0 else 'green' for x in df['mpg_z']]
df.sort_values('mpg_z', inplace=True)
df.reset_index(inplace=True)

# Draw plot
plt.figure(figsize=(14,10), dpi= 80)
plt.hlines(y=df.index, xmin=0, xmax=df.mpg_z, color=df.colors, alpha=0.4, linewidth=5)

#Plot styling 
plt.gca().set(ylabel='Model', xlabel='Mileage by standard deviation')
plt.yticks(df.index, df.cars, fontsize=12)
plt.title('Millage my STD', fontdict={'size':20})
plt.grid(linestyle='--', alpha=0.5)
```

```python
#loading dataset. 
df = pd.read_csv("mtcars.csv")

#Configuring data for heatmap 
plt.figure(figsize=(12,10), dpi= 70)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='Blues', 
            center=0, annot=True)

# Configuring legend/title/fonts
plt.title('Correlation of columns in dataset', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

#NOTE: Compiling this in Jupyter notebook cuts off 1/2 of the top and bottom row of this visualization 
```
